1. Write a program that asks the user for the number of days. The program then prints out the number of seconds in the number of days given.

```c++
#include <iostream>
using namespace std;

{
    int days;
    const int SECONDS_IN_A_DAY = 86400; // 24 x 60 x 60

    cout << "Enter no. of days: ";
    cin >> days;

    // Cast to long long — prevents integer overflow for large day values
    long long total_seconds = (long long)days * SECONDS_IN_A_DAY;

    cout << days << " days is exactly " << total_seconds << " seconds." << endl;
}
```

2. Write a program that asks a user to input the radius then the program calculates the volume of a sphere (the formula for the volume is (4/3) πr3). Use the inbuilt exponential function in C++ to compute (r3).

```c++
#include <iostream>
#include <cmath>    

using namespace std;

double radius;

// Ask user to enter the radius
cout << "Enter the radius of the sphere: ";
cin  >> radius;

// Validate: radius cannot be negative
if (radius < 0) {
    cout << "Error: radius must be a positive value." << endl;
} else {
    // Compute r³ using the inbuilt pow() function
    double rCubed = pow(radius, 3);

    // Apply the sphere volume formula: V = (4/3) × π × r³
    // M_PI is a constant for 3.14...
    double volume = (4.0 / 3.0) * M_PI * rCubed;

    // Display the computed values
    cout << "r³ (via pow) = " << rCubed << endl;
    cout << "Volume       = " << volume << endl;
}
```

3. Using functions, write a program to compute the area and perimeter of a square.
The program should ask the user to enter a number corresponding to the side length
of the square and display the area and perimeter of the square.


```c++
#include <iostream>
using namespace std;

// Function: returns area of a square given side length
double computeArea(double side) {
    return side * side;
}

// Returns perimeter of a square given side length
double computePerimeter(double side) {
    return 4 * side;
}

double side;

// Prompt the user to enter the side length
cout << "Enter the side length of the square: ";
cin >> side;

// Call functions and display results
cout << "Area      = " << computeArea(side) << endl;
cout << "Perimeter = " << computePerimeter(side) << endl;
```

4. Write a program using functions that determines whether a character input by a
user is uppercase or lower case.

```c++
#include <iostream>
#include <cctype>   // provides isupper() and islower()
using namespace std;

// Function: checks and prints whether ch is upper, lower, or neither
void checkCase(char ch) {
    if (isupper(ch)) {
        cout << "'" << ch << "' is an UPPERCASE letter." << endl;
    } else if (islower(ch)) {
        cout << "'" << ch << "' is a lowercase letter." << endl;
    } else {
        // Character is a digit, symbol, or whitespace
        cout << "'" << ch << "' is not a letter." << endl;
    }
}

char ch;

// Ask the user for a single character
cout << "Enter a character: ";
cin >> ch;

checkCase(ch);   // delegate logic to function
```

5. The following is pseudocode for a program being designed. Write the C++ program
equivalent of the same.
BEGIN
    SET x TO 0, y TO 20
    REPEAT
    SUBTRACT 4 FROM y
    ADD 2/y TO x
    UNTIL
    y IS LESS THAN 6
    DISPLAY x
END


```c++
#include <iostream>
using namespace std;

double x = 0;
double y = 20;

// do-while mirrors the REPEAT...UNTIL structure:
// the body runs first, then the condition is checked
do {
    y -= 4;              // SUBTRACT 4 FROM y
    x += 2.0 / y;        // ADD 2/y TO x (2.0 ensures floating-point division)
} while (y >= 6);       // loop continues UNTIL y < 6

// Display the final value of x
cout << "x = " << x << endl;
```

6. Write a C++ program that does the following:
i. Uses a loop for a user to continually input 5 values to populate an array.
ii. Calculates and displays the average of the values input into the array.


```c++
#include <iostream>
using namespace std;

const int SIZE = 5;        // fixed array size
double values[SIZE];       // declare array to hold 5 values
double sum = 0;

// Loop to collect 5 values from the user
for (int i = 0; i < SIZE; i++) {
    cout << "Enter value " << (i + 1) << ": ";
    cin >> values[i];        // store input into array at index i
    sum += values[i];        // accumulate running total
}

// Compute the average by dividing the sum by the array size
double average = sum / SIZE;

cout << "Average = " << average << endl;
```
